```python
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen en escala de grises
imagen = cv2.imread('perro.jpg', cv2.IMREAD_GRAYSCALE)
#Mostrar imagen con OpenCV (requiere ventana emergente)
cv2.imshow('Imagen', imagen)
cv2.waitKey(0)
cv2.destroyAllWindows()
#Mostrar con Matplotlib (evita problemas con colores)
plt.imshow(imagen, cmap='gray')
plt.axis('off')
plt.show()
```

```python
bordes = cv2.Canny(imagen, 100, 200)
plt.imshow(bordes, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Rotacin de 45 grados alrededor del centro
(h, w) = imagen.shape[:2]
M = cv2.getRotationMatrix2D((w//2, h//2), 45, 1.0)
imagen_rotada = cv2.warpAffine(imagen, M, (w, h))
```

```python
plt.hist(imagen_ecualizada, bins=20)
plt.plot()
```

```python
# Filtro Gaussiano
imagen_suavizada = cv2.GaussianBlur(imagen, (5,5), 0)
# Filtro de mediana
imagen_median = cv2.medianBlur(imagen, 5) 
```

```python
plt.imshow(imagen_suavizada, cmap='gray')
plt.axis('off')
plt.show()
```
